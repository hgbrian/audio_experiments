Test loss and grad of a 1 pixel image.

```python
import numpy as np
import theano
import theano.tensor as T
from IPython.display import SVG, HTML

def test_loss(P, X):
    loss = 1./2 * ((X - P)**2).sum()
    return loss

# Photo
P = theano.shared(np.array([1,1,1], np.float32))

# Generated image
X = theano.shared(np.array([3,0,4]).astype(theano.config.floatX))

total_loss = test_loss(P, X)
grad = T.grad(total_loss, X)
test_f_grad = theano.function([], grad)

print "P:", P, P.eval()
print "X:", X, X.eval()
print "total_loss", total_loss
print "test_grad", grad

print "test_f_grad with X=[3,0,4]: ", np.array(test_f_grad())
X.set_value([3,10,4])
print "test_f_grad with X=[3,10,4]:", np.array(test_f_grad())

HTML("<hr><h1>Pre-compilation graph</h1>" + \
     theano.printing.pydotprint(grad, return_image=True, format='svg', var_with_name_simple=True) + \
     "<hr><h1>Post-compilation graph</h1>" + \
     theano.printing.pydotprint(test_f_grad, return_image=True, format='svg', var_with_name_simple=True))
```
